```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
X = [[174],[152],[138],[128],[186]]
y = [71, 55, 46, 38, 88]
model.fit(X, y)
```

```python
model.coef_ # y = ax +b 에서 a에 속하는 값
```

```python
model.intercept_ # y = ax +b 에서 b에 속하는 값
```

```python
model.predict([[165]])
```

```python
model2 = LinearRegression()

X = [[0], [1], [2]]
Y = [3,4,5]

model2.fit(X,Y)

print(model2.coef_)
print(model2.intercept_)

print(model2.predict([[3]]))
```

```python
from matplotlib import pyplot as plt
y_hat = model.predict(X)

plt.scatter(X, y)
plt.plot(X, y_hat, 'r--')
plt.show()
```

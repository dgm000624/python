```python
import numpy as np
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7,8,9]
y = map(lambda x:x**2, x)
print(list(y))
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(-20,20)
y1 = 2*x
y2 = (1/3)*x**2+5
y3 = -(x**2)-5
plt.plot(x,y1,'r^-',x,y2,'g*',x,y3,'b.:')
plt.axis([-30,30,-30,30])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel("ForCal.xlsx")
fig = plt.figure()

df = df[::-1]
# df['date'] = pd.to_datetime(df['date'])
# df = df.set_index('date')
x = np.array(df['date'][::10])
y1 = df['won/dollar'][::10]
plt.plot(x, y1, 'b', label = 'won/dollar')
plt.xlabel('period')
plt.ylabel('percent')
plt.plot(x, df['gold_price'][::10], 'y--', label = 'gold_price')
plt.plot(x, df['VIX_종가'][::10], 'r', label = 'VIX_Close_price')
plt.plot(x, df['금ETF'][::10], label = 'Gold_ETF')
plt.legend(loc='best')

fig.savefig('all_visual.png')
plt.show()

```

```python
x = np.linspace(0, np.pi*2, 100)
fig = plt.figure()

plt.plot(x, np.sin(x), 'r--', label="sin")
plt.plot(x, np.cos(x), 'g', label = 'cos')

plt.xlabel('6 = 2pi')
plt.ylabel('green: cos, red dot_line: sin')

plt.title('sin & cos wave')
plt.suptitle('grd')
plt.legend(loc='best')

fig.savefig('sin&cos.png')
plt.show()
```
